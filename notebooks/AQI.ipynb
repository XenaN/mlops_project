```python
df_aqi = pd.read_csv("../data/processed/AQI_dataset.csv", parse_dates=['Datetime'], index_col=['Datetime'])
fig = plt.figure(figsize=(15, 10))
df_aqi["AQI"].plot()
```

```python
df_aqi_daily = df_aqi.resample('D').mean().interpolate()
fig = plt.figure(figsize=(15, 10))
df_aqi_daily["AQI"].plot()
```

***Time series analysis***

Stationarity

```python
from statsmodels.tsa.stattools import adfuller
ADF_result = adfuller(df_aqi_daily["AQI"])   
print(f'ADF Statistic: {ADF_result[0]}') 
print(f'p-value: {ADF_result[1]}')
```

With a large negative ADF statistic and a p-value much smaller than 0.05, we can say that our series is stationary.

Autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df_aqi_daily["AQI"], lags=50)
plt.tight_layout()
```

```python
plot_pacf(df_aqi_daily["AQI"], lags=50)
plt.show()
```

Looks like autoregressive process 

```python
from statsmodels.tsa.seasonal import seasonal_decompose

analysis = df_aqi_daily["AQI"].copy()


decompose_result_mult = seasonal_decompose(analysis, model="multiplicative")

trend = decompose_result_mult.trend
seasonal = decompose_result_mult.seasonal
residual = decompose_result_mult.resid

fig = plt.figure(figsize=(15, 10))
seasonal[:100].plot();
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df_aqi_daily["AQI"], test_size=0.2, shuffle=False) 
```

```python
model = ARIMA(train, order=(2,0,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
predictions_p_ms = model_fit.predict(start=len(train), end=len(train)+5)
```

```python
arima_test_rmse = mean_squared_error(test.iloc[:6], predictions_p_ms, squared=False)
print('Test RMSE: %.3f' % arima_test_rmse)
```

```python
rmse_arima = []
n = 1
# train = train.values
for i in range(len(test)-n):
    model = ARIMA(train, order=(4,0,0))
    model_fit = model.fit()
    predictions_p_ms = model_fit.predict(start=len(train), end=len(train)+n)
    arima_test_rmse = mean_squared_error(test.iloc[i:i+n+1], predictions_p_ms, squared=False)
    rmse_arima.append(arima_test_rmse)
    
    train = np.append(train, test.iloc[0])

print("RMSE: {}".format(np.mean(rmse_arima)))
```

```python
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test.iloc[:6], label='actual')
plt.plot(predictions_p_ms, label='forecast')
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
train_size = int(df_naive.shape[0] * 0.8)
train, test = df_naive[1:train_size], df_naive[train_size:]
train_X, train_y = train["t-1"], train["t+1"]
test_X, test_y = test["t-1"], test["t+1"]
```

```python
model_forest = RandomForestRegressor(n_estimators=1000)
model_forest.fit(train_X.values.reshape(-1, 1), train_y.values)

y_hat = model_forest.predict(np.array(test_X)[:, np.newaxis])

forest_test_rmse = mean_squared_error(test_y, y_hat, squared=False)  
```

```python
print(forest_test_rmse)
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train["t-1"], label='training')
plt.plot(test["t-1"], label='actual')
plt.plot(test.index, y_hat, label='forecast')
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
tscv = TimeSeriesSplit()
TimeSeriesSplit(max_train_size=None, n_splits=5)
rmse = []
for train_index, test_index in tscv.split(df_naive):
    train_cv, test_cv = df_naive.iloc[train_index], df_naive.iloc[test_index]
    train_X_cv, train_y_cv = train_cv["t-1"], train_cv["t+1"]
    test_X_cv, test_y_cv = test_cv["t-1"], test_cv["t+1"]  
    
    model_forest_cv = RandomForestRegressor(n_estimators=1000)
    model_forest_cv.fit(train_X_cv.values.reshape(-1, 1), train_y_cv.values)

    y_hat_cv = model_forest_cv.predict(np.array(test_X_cv)[:, np.newaxis])

    rmse_cv = mean_squared_error(test_y_cv, y_hat_cv, squared=False)  
    rmse.append(rmse_cv)
    
    plt.figure(figsize=(12,5), dpi=100)
    plt.plot(train_cv["t-1"], label='training')
    plt.plot(test_cv["t-1"], label='actual')
    plt.plot(test_X_cv.index, y_hat_cv, label='forecast')
    plt.title('Forecast vs Actuals')
    plt.legend(loc='upper left', fontsize=8)
    plt.show()

print("RMSE: {}".format(np.mean(rmse)))
```

```python
for depth in range(1, 16):
    y = df_aqi_daily['AQI'].copy()
    x_list = [df_aqi_daily[cols].shift(i) for i in range(1, depth+1)]
    x = pd.concat(x_list, axis=1)

    x = x.iloc[depth:]
    y = y.iloc[depth:]

    X_train, X_test, y_train, y_test = train_test_split(x, y, shuffle=False, random_state=42)
    model_forest = RandomForestRegressor(**{'n_estimators': 585,
     'min_samples_split': 2,
     'min_samples_leaf': 8,
     'max_depth': 52,
     'bootstrap': True})

    model_forest.fit(X_train, y_train)

    y_hat = model_forest.predict(X_test)

    forest_test_rmse = mean_squared_error(y_test, y_hat, squared=False) 
    print(depth, forest_test_rmse)
```
